```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/test.csv')

# View first 5 rows
df.head()

```

```python
# Column names and data types
df.info()
```

```python
# Shape of dataset
df.shape
```

```python
# Statistical summary
df.describe(include='all')
```

```python
#1. Handling Missing Values
#a) Check Missing Values
df.isnull().sum()
```

```python
#b) Fill Missing Values with Mean (Numerical)
df['Age'] = df['Age'].fillna(df['Age'].mean())
print(df['Age'])
```

```python
#c)Median Imputation (Numerical: Flight Distance)
df['Flight Distance'] = df['Flight Distance'].fillna(df['Flight Distance'].median())
print(df['Flight Distance'])
```

```python
#d)Mode Imputation (Categorical: gender)
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
print(df['Gender'])
```

```python
#e)Drop Rows with Missing Text ()
df.dropna(subset=['Inflight wifi service'], inplace=True)
print(df['Inflight wifi service'])
```

```python
# 2. Handling Duplicate Records

# Check Duplicates
df.duplicated().sum()

# Remove Duplicates
df.drop_duplicates(inplace=True)

# Show duplicate rows
duplicates = df[df.duplicated()]
print(duplicates)
```

```python
# 3. Detecting & Handling Outliers
import seaborn as sns
import matplotlib.pyplot as plt

sns.boxplot(x=df['Flight Distance'])
plt.show()
```

```python
# 4. Handling Inconsistent Data
# a) Fix Gender Values
df['gender'] = df['Gender'].str.lower().str.strip()

df['gender'] = df['Gender'].replace({
    'm': 'male',
    'f': 'female'
})
print(df)
```

```python
# 5.Renaming Columns
df.rename(columns={
    'Type of Travel': 'Type_of_Travel',
    'Flight Distance': 'Flight_Distance'
}, inplace=True)
print(df)
```

```python
#6. Data Type Conversion
df['Flight_Distance'] = df['Flight_Distance'].astype(int)
print(df)
```

```python
#7 Scaling Data (Standardization)
df['Flight_Distance_Zscore'] = (
    df['Flight_Distance'] - df['Flight_Distance'].mean()
) / df['Flight_Distance'].std()

df[['Flight_Distance', 'Flight_Distance_Zscore']].head()
```

```python
#8. Binning
bins = [0, 18, 30, 45, 60, 100]
labels = ['Teen', 'Young Adult', 'Adult', 'Senior', 'Old']

df['Age_Group'] = pd.cut(df['Age'], bins=bins, labels=labels)

df[['Age', 'Age_Group']].head()
```

```python
#9. Aggregation
# Average Age by Customer Type
df.groupby('Customer Type')['Age'].mean()
print(df)
```

```python
#19. EDA (Exptoratory data analysis)
#Mean, Median, Mode
print("Mean Age:", df['Age'].mean())
print("Median Age:", df['Age'].median())
print("Mode Age:", df['Age'].mode()[0])
```

```python
#11. Box Plot (Visualization Output)
# Flight Distance
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Flight_Distance'])
plt.title("Box Plot of Flight Distance")
plt.show()
```

```python
#12. Histogram (WITH DISPLAY)
# Age
plt.figure(figsize=(6,4))
plt.hist(df['Age'], bins=20)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution")
plt.show()
```

```python
#13. Bar Chart (WITH PRINT)
# Satisfaction Count
print("Satisfaction Count:")
print(df['satisfaction'].value_counts())

df['satisfaction'].value_counts().plot(kind='bar')
plt.title("Customer Satisfaction Count")
plt.xlabel("Satisfaction")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
#14. Scatter Plot (WITH VISUAL OUTPUT)
# Flight Distance vs Arrival Delay
plt.figure(figsize=(6,4))
plt.scatter(df['Flight_Distance'], df['Arrival Delay in Minutes'], alpha=0.5)
plt.xlabel("Flight Distance")
plt.ylabel("Arrival Delay (Minutes)")
plt.title("Flight Distance vs Arrival Delay")
plt.show()
```

```python
#15. Feature Engineering
#One-Hot Encoding
class_ohe = pd.get_dummies(df['Class'], prefix='Class')

print("One-Hot Encoded Class (first 5 rows):")
print(class_ohe.head())
df = pd.concat([df, class_ohe], axis=1)
```

```python
#16. Dummy Variable Creation (Type_of_Travel)
travel_dummy = pd.get_dummies(df['Type_of_Travel'], drop_first=True)

print("Dummy Variables for Type_of_Travel:")
print(travel_dummy.head())
df = pd.concat([df, travel_dummy], axis=1)
```

```python
#17. Label Encoding (satisfaction)
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()   # <-- missing line

df['Satisfaction_Label'] = le.fit_transform(df['satisfaction'])

print("Satisfaction Label Encoding:")
print(df[['satisfaction', 'Satisfaction_Label']].head())
```

```python
#18. Feature Extraction
# Total Delay Feature
df['Total_Delay'] = (
    df['Departure Delay in Minutes'] +
    df['Arrival Delay in Minutes']
)

print("Total Delay Feature:")
print(df[['Departure Delay in Minutes', 'Arrival Delay in Minutes', 'Total_Delay']].head())
```

```python
#19. Feature Scaling
# Standard Scaling
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaled_features = scaler.fit_transform(
    df[['Age', 'Flight_Distance', 'Total_Delay']]
)

scaled_df = pd.DataFrame(
    scaled_features,
    columns=['Age_Scaled', 'Flight_Distance_Scaled', 'Total_Delay_Scaled']
)

print("Scaled Features:")
print(scaled_df.head())
df = pd.concat([df, scaled_df], axis=1)
```

```python
#20 Dimensionality Reduction (PCA)
from sklearn.impute import SimpleImputer
from sklearn.decomposition import PCA

imputer = SimpleImputer(strategy='mean')

X_imputed = imputer.fit_transform(
    df[['Age_Scaled', 'Flight_Distance_Scaled', 'Total_Delay_Scaled']]
)

pca = PCA(n_components=2)
pca_features = pca.fit_transform(X_imputed)

pca_df = pd.DataFrame(
    pca_features,
    columns=['PCA_Component_1', 'PCA_Component_2']
)

df = pd.concat([df.reset_index(drop=True), pca_df], axis=1)
print(df[['PCA_Component_1', 'PCA_Component_2']].head())
```

```python
#21. TEXT DATA PROCESSING TECHNIQUES
# Lowercasing
df['satisfaction_lower'] = df['satisfaction'].str.lower()

print("Lowercasing Output:")
print(df[['satisfaction', 'satisfaction_lower']].head())
```

```python
df.to_csv("final_processed_airline_data.csv", index=False)
print("Final processed dataset saved successfully!")
```
